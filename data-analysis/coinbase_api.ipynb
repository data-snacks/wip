## Currencies

Currencies API has two endpoints - currencies (flat currencies) and crypto currencies https://docs.cdp.coinbase.com/sign-in-with-coinbase/docs/api-currencies/

```python
currencies = requests.get('https://api.coinbase.com/v2/currencies') #Get Fiat Currencies
crypto_currencies = requests.get("https://api.coinbase.com/v2/currencies/crypto") # Get Cryptocurrencies
```

### Flat

```python
currencies.json()
```

## Crypto

# Prices

https://docs.cdp.coinbase.com/sign-in-with-coinbase/docs/api-prices/

```python
bare_url = 'https://api.coinbase.com/v2/prices/BTC-USD/'

def get_prices(root_url):
    for operation in ["buy", "sell", "spot"]:
        full_url = root_url + operation
        request = requests.get(full_url)
        print(operation, '->' ,request.text)
```

```python
get_prices(bare_url)
```
